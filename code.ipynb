```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("data2.csv")
```

```python
df.head(10)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
```

```python
# Splitting the data into features (X) and target (y)
X = df.drop('MILLET', axis=1)
y = df['MILLET']
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X.columns
```

```python
y
```

```python
# Initialize the Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Training the model
model.fit(X_scaled, y)
```

```python
# Making predictions on the test set
y_pred = model.predict(X_scaled)
```

```python
# Calculating accuracy
accuracy = accuracy_score(y, y_pred)
print(f"Accuracy of the RFC model: {accuracy:.2f}")
```

```python
# Splitting the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.4, random_state=42)
```

```python
X_test
```

```python
# Initialize the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Training the model
model2.fit(X_train, y_train)
```

```python
# Making predictions on the test set
y_pred2 = model2.predict(X_test)
```

```python
# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred2)
print(f"Accuracy of the RFC model: {accuracy:.2f}")
```

```python
# Creating a correlation matrix
corr_matrix = df.corr()
print("Correlation Matrix:")
print(corr_matrix)

# Visualizing the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scale the user input using the same scaler
user_input = scaler.transform([[protein, carbs, fat, minerals, fiber, calcium, phosphorus, iron, energy, thiamin, niacin]])
predicted_crop = model2.predict(user_input)

print(f"The predicted crop based on the input is: {predicted_crop[0]}")
```

```python
# Scale the user input using the same scaler
user_input = scaler.transform([[protein, carbs, fat, minerals, fiber, calcium, phosphorus, iron, energy, thiamin, niacin]])

# Predict probabilities for each class
predicted_probabilities = model.predict_proba(user_input)[0]

# Get the indices of the top three predicted classes
top_3_indices = predicted_probabilities.argsort()[-3:][::-1]

print("Top 3 predicted classes:")
for index in top_3_indices:
    print(f"{model.classes_[index]}: {predicted_probabilities[index] * 100:.2f}% probability")
```
